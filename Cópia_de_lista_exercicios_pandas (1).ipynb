## 2. Seleção e Filtragem

### Questão 1

Filtre registros com valores acima da média em uma coluna numérica.

```python
media = df['length_url'].mean()
media
```

```python
media_filtrada = df[df['length_url'] > media]
media_filtrada.sort_values('length_url')
```

### Questão 2

Filtre registros pertencentes a uma categoria específica.

```python
df_filtrado = df[df['ip'] == 1]
df_filtrado
```

### Questão 3

Filtre registros com múltiplas condições.

```python
registro_filtrado = df[(df['length_url'] < 200) & (df['length_url'] > 50)]
registro_filtrado
```
